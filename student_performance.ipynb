```python
def fillnaObjectMode(cols):
    for col in cols:
        df[col] = df[col].fillna(df[col].mode()[0])

object_cols = df.select_dtypes(include="object").columns.tolist()
fillnaObjectMode(object_cols)

```

```python
def fillnaIntMode(cols):
    for col in cols:
        df[col] = df[col].fillna(df[col].mode()[0])

int_cols = df.select_dtypes(include="int64").columns.tolist()
fillnaIntMode(int_cols)


```

```python
def fillnaFloatMean(cols):
    for col in cols:
        df[col] = df[col].fillna(df[col].mean())

float_cols = df.select_dtypes(include="float64").columns.tolist()
fillnaFloatMean(float_cols)

df.isnull().sum()
df.info()
```

```python
def changeFloatToInt(cols):
    for col in cols:
        df[col] = df[col].astype("int64")

float_cols = df.select_dtypes(include="float64").columns.tolist()
changeFloatToInt(float_cols)

df.info()
```

```python


def convertObjToInt(cols):
    for col in cols:
        unique_vals = pd.DataFrame({col: df[col].unique()})
        le = LabelEncoder()
        le.fit(np.ravel(unique_vals))
        df[col] = le.transform(df[col])

object_cols = df.select_dtypes(include="object").columns.tolist()
convertObjToInt(object_cols)

```

```python
df.info()
```

```python
df.to_csv("processed_data.csv", index=False)
```

```python
df = pd.read_csv("processed_data.csv")
df
```

```python


X = df.drop("Pass_Fail", axis=1)  
y = df["Pass_Fail"]               

train_X, test_X, train_y, test_y = train_test_split(
    X, y, test_size=0.3, shuffle=False
)

print(train_X.shape, train_y.shape)
print(test_X.shape, test_y.shape)
```

```python
from sklearn.svm import SVC

model_svc = SVC()
model_svc.fit(train_X, train_y)

print(model_svc)
```

```python
model_pred = model_svc.predict(test_X)
model_pred
```

```python


acc = accuracy_score(test_y, model_pred)
print("Accuracy:", round(acc * 100, 2), "%")
```
